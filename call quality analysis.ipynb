```python
# def runknown(oprtr,ntwrk):
#     if ntwrk=='Unknown':
#         if oprtr=='Airtel' or oprtr=='RJio':
#             ntwrk='4G'
#         elif oprtr=='BSNL' or oprtr=='Idea' or oprtr=='Tata' or oprtr=='Vodafone':
#             ntwrk='3G'
#         else:
#             ntwrk='2G'
    
newapril=april
newapril.loc[((newapril["Operator"] == 'Airtel') |  (newapril["Operator"] == 'RJio')) & (newapril["Network Type"]=='Unknown'), "Network Type"] = '4G'  
newapril.loc[((newapril["Operator"] == 'MTNL') |  (newapril["Operator"] == 'Other')) & (newapril["Network Type"]=='Unknown'), "Network Type"] = '2G'  
newapril.loc[(newapril["Network Type"]=='Unknown'), "Network Type"] = '2G'  

newapril.fillna("",inplace=True)
newapril
```

```python
    
newmay=may
newmay.loc[((newmay["Operator"] == 'Airtel') |  (newmay["Operator"] == 'RJio')) & (newmay["Network Type"]=='Unknown'), "Network Type"] = '4G'  
newmay.loc[((newmay["Operator"] == 'MTNL') |  (newmay["Operator"] == 'Other')) & (newmay["Network Type"]=='Unknown'), "Network Type"] = '2G'  
newmay.loc[(newmay["Network Type"]=='Unknown'), "Network Type"] = '2G'  

newmay.fillna("",inplace=True)
newmay
```

```python
print(newapril.columns)
print(newmay.columns)
```

```python
aprilme=newapril
# aprilme["Network"]=newapril["Operator"]+" "+newapril["Network Type"]
aprilme.drop(["Latitude","Longitude"],axis=1,inplace=True)

mayme=newmay
mayme.drop(["Latitude","Longitude"],axis=1,inplace=True)
# mayme["Network"]=newmay["Operator"]+" "+newmay["Network Type"]
```

```python
# def columns(Dataframe):
#     df=Dataframe
#     cols=list(df.columns)
#     cols=[cols[-1]]+cols[:-1]
#     return df[cols]

# aprilme=columns(aprilme)
aprilme.columns=['Operator', 'In Out Travelling', 'Network Type', 'Rating','Call Drop Category','State Name']
mayme.columns=['Operator', 'In Out Travelling', 'Network Type', 'Rating','Call Drop Category','State Name']
aprilme["Month"]="April"
mayme["Month"]="May"
```

```python
callqlty=pd.concat([aprilme,mayme])
callqlty
```
